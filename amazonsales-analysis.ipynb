## Changing Data Types from object to float

```python
# Changing the data type of discounted price and actual price

df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

## Statistical Test: T-test

### Do highly discounted products have different ratings than low-discount products?
### H0: Mean rating is the same
### H1: Mean rating is different

```python
from scipy.stats import ttest_ind

high_discount = df[df['discount_percentage'] > df['discount_percentage'].median()]['rating']
low_discount = df[df['discount_percentage'] <= df['discount_percentage'].median()]['rating']

t_stat, p_value = ttest_ind(high_discount, low_discount)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

Insight: An independent samples t-test was conducted to determine whether customer ratings differed between highly discounted and low-discount products. The results showed a statistically significant difference in mean ratings between the two groups (t = -4.21, p < 0.001). This suggests that discount level is associated with differences in customer ratings, although the direction and practical significance should be interpreted alongside the group means. Therefore, the null hypothesis was rejected, providing evidence that discount level is associated with differences in customer ratings.

While the correlation between discount percentage and rating was weak (r = -0.15), the t-test revealed a statistically significant difference in average ratings between high-discount and low-discount products. This suggests that discount levels may influence ratings, although the effect size appears to be small.

# Insights

### Consumer Satisfaction and Pricing

The correlation analysis revealed that product price has very little influence on customer ratings. The Pearson correlation coefficient between actual price and rating was 0.12, indicating a very weak positive relationship. This suggests that consumers do not necessarily perceive higher-priced products as being of better quality or deserving of higher ratings. Instead, customer satisfaction appears to be driven by factors other than price, such as product performance, reliability, and overall user experience. From a consumer analytics perspective, this finding indicates that premium pricing alone is unlikely to improve customer perceptions or satisfaction levels.

### Consumer Response to Discounts

The relationship between discount percentage and customer ratings was found to be weakly negative (r = -0.15), suggesting that products with larger discounts tend to receive slightly lower ratings. To further investigate this relationship, an independent samples t-test was conducted, which revealed a statistically significant difference in ratings between high-discount and low-discount products (t = -4.21, p < 0.001). This indicates that discount levels are associated with differences in customer satisfaction. However, the weak correlation suggests that while discounts may influence purchasing decisions, they are not a major driver of customer ratings. Consumers appear to place greater emphasis on product quality and value rather than promotional pricing alone.

### Consumer Engagement and Product Popularity

The analysis also examined the relationship between customer ratings and rating counts as a measure of product popularity and engagement. A very weak positive correlation (r = 0.10) was observed between ratings and rating counts, indicating that products with more reviews tend to have slightly higher ratings. However, the relationship is minimal, suggesting that popularity does not necessarily translate into higher customer satisfaction. This finding implies that consumers continue to evaluate products based on their individual experiences rather than being strongly influenced by the volume of existing reviews.

### Pricing Structure and Market Positioning

A very strong positive correlation (r = 0.96) was identified between actual price and discounted price. This indicates that products with higher original prices generally remain more expensive after discounts are applied. The result suggests that discounts primarily serve as a pricing adjustment mechanism rather than fundamentally altering a product's market position. Premium products continue to occupy premium price segments even during promotional periods. This finding highlights how retailers use discounts to encourage purchases while maintaining established product positioning and perceived value among consumers.

### Overall Consumer Analytics Insight

Overall, the findings suggest that customer satisfaction is largely independent of pricing and discount strategies. While discounts may create statistically significant differences in ratings, the overall relationships between price, discounts, and customer ratings remain weak. Consumers appear to evaluate products primarily on their quality, functionality, and overall experience rather than on price alone. These results indicate that businesses seeking to improve customer satisfaction should focus on enhancing product performance and customer experience rather than relying solely on pricing promotions or discount strategies.
